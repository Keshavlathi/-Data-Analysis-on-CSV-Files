```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv("/content/Sample - Superstore.csv", encoding='latin-1')
```

```python
print("\n===== First 5 Rows =====")
display(sales.head())
```

```python

print("\n===== Dataset Shape =====")
print(f"Rows: {sales.shape[0]}, Columns: {sales.shape[1]}")

```

```python

print("\n===== Info =====")
display(sales.info())
```

```python
print("\n===== Missing Values =====")
display(sales.isnull().sum())
```

```python
print("\n===== Duplicate Records =====")
print(sales.duplicated().sum())
```

```python
display(sales.describe())
```

```python
category_sales = sales.groupby('Category')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
sb.barplot(x=category_sales.values, y=category_sales.index, palette="Blues_r")
plt.title('💰 Total Sales by Category', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Sales ($)', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.show()
```

```python
region_sales = sales.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
sb.barplot(x=region_sales.values, y=region_sales.index, palette="Oranges_r")
plt.title('🌍 Total Sales by Region', fontsize=16, fontweight='bold', color='darkred')
plt.xlabel('Sales ($)', fontsize=12)
plt.ylabel('Region', fontsize=12)
plt.show()
```

```python
ship_modes = sales['Ship Mode'].value_counts()
plt.figure(figsize=(6,4))
sb.barplot(x=ship_modes.values, y=ship_modes.index, palette="Purples_r")
plt.title('🚚 Orders by Shipping Mode', fontsize=16, fontweight='bold', color='purple')
plt.xlabel('Number of Orders', fontsize=12)
plt.ylabel('Shipping Mode', fontsize=12)
plt.show()
```

```python
top_product = sales.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,5))
sb.barplot(x=top_product.values, y=top_product.index, palette="Greens_r")
plt.title('🏆 Top 5 Selling Products', fontsize=16, fontweight='bold', color='green')
plt.xlabel('Quantity Sold', fontsize=12)
plt.ylabel('Product Name', fontsize=12)
plt.show()
```

```python

print("\nTop Selling Product:")
display(top_product.head(1))
```
